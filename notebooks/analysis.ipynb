```python
#  Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#  Load Data
historical_demand = pd.read_csv("../data/historical_demand.csv")
staff_details = pd.read_csv("../data/staff_details.csv")
hiring_firing_costs = pd.read_csv("../data/hiring_firing_costs.csv")
```

```python
print("Historical Demand Data:\n", historical_demand.head())
print("\nStaff Details:\n", staff_details.head())
print("\nHiring & Firing Costs:\n", hiring_firing_costs.head())
```

```python
# Plot Demand Trends
plt.figure(figsize=(12, 6))
sns.lineplot(x=historical_demand["Month"], y=historical_demand["Morning_C"], label="Morning - Caregivers")
sns.lineplot(x=historical_demand["Month"], y=historical_demand["Evening_C"], label="Evening - Caregivers")
sns.lineplot(x=historical_demand["Month"], y=historical_demand["Night_C"], label="Night - Caregivers")
plt.title("Staffing Demand Trends (Caregivers)")
plt.xlabel("Month")
plt.ylabel("Required Staff")
plt.legend()
plt.grid()
plt.show()
```

```python
#  Hiring vs. Firing Cost Impact
plt.figure(figsize=(8, 4))
sns.barplot(x=hiring_firing_costs["Staff Category"], y=hiring_firing_costs["Hiring Cost (?)"], color="blue", label="Hiring Cost")
sns.barplot(x=hiring_firing_costs["Staff Category"], y=hiring_firing_costs["Firing Cost (?)"], color="red", label="Firing Cost")
plt.title("Hiring vs. Firing Cost Comparison")
plt.ylabel("Cost ()")
plt.legend()
plt.show()
```

```python
#  Analyzing Cost Impact
staff_costs = staff_details.copy()
staff_costs["Total Cost (Weekly)"] = staff_costs["Hourly Wage (?)"] * staff_costs["Regular Hours/Week"]
print("\n Estimated Weekly Cost per Staff Category:\n", staff_costs[["Staff Category", "Total Cost (Weekly)"]])
```

```python
#  Conclusion
print("\n Insights:")
print("- Peak staffing demand occurs in December & January.")
print("- Night shifts require better distribution to reduce burnout.")
print("- Hiring costs are significantly higher than firing costs for nurses.")
```
